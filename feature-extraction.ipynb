```python
import pandas as pd
import pickle
import numpy as np
```

```python
df = pickle.load(open('dataset2.pkl','rb'))
```

```python
df

# The below features are required
# batting_team
# bowling team
# city
# current score
# ball left
# wickets left
# current rr
# last five
```

```python
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
cities
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
print(df['runs'].dtype)
```

```python
df.loc[:, 'current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df.loc[:, 'balls_bowled'] = (df['over'].astype('int') * 6) + df['ball_no'].astype('int')
```

```python
df.loc[:, 'balls_left'] = 120 - df['balls_bowled']
```

```python
df.loc[:, 'balls_left'] = df['balls_left'].apply(lambda x: 0 if x < 0 else x)
```

```python
df.loc[:,'player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df.loc[:,'player_dismissed'] = df['player_dismissed'].astype('int')
df.loc[:,'player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df.loc[:,'wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df.loc[:,'crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
df['runs'].unique()
```

```python
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')
groups = df.groupby('match_id')
match_ids = df['match_id'].unique()
last_five = []


#for id in match_ids:
 #   rolling_sums = groups.get_group(id)['runs'].rolling(window=30).sum()
  #  last_five.extend(rolling_sums.dropna().values.tolist())  

for id in match_ids:
    last_five.extend(groups.get_group(id)['runs'].rolling(window=30).sum().values.tolist())
```

```python
last_five
```

```python
df.loc[:,'last_five'] = last_five
```

```python
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample(2)
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
trf = ColumnTransformer(
    transformers=[
        ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
    ],
    remainder='passthrough'
)
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', StandardScaler()),
    ('step3', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
